```python
# retrieve data from PDF
import pdfplumber
import re
import pandas as pd
import numpy as np

def extract_invoice_data(pdf_path, output_excel=None):
    """
    Extract specific fields from an unstructured PDF (e.g., invoices).

    Parameters:
        pdf_path (str): Path to the PDF file.
        output_excel (str, optional): Path to save the extracted data as an Excel file.

    Returns:
        pd.DataFrame: Extracted data in a structured format.
    """
    extracted_data = []

    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page_number, page in enumerate(pdf.pages, start=1):
                text = page.extract_text()
                if text:
                    # Example: Extract fields using regex
                    invoice_number = re.search(r"Invoice Number[:\\s]*([A-Za-z0-9-]+)", text)
                    date = re.search(r"Date[:\\s]*([0-9/.-]+)", text)
                    total = re.search(r"Total[:\\s]*\\\\$?([0-9,]+\\.[0-9]{2})", text)

                    extracted_data.append({
                        "Page": page_number,
                        "Invoice Number": invoice_number.group(1) if invoice_number else None,
                        "Date": date.group(1) if date else None,
                        "Total": total.group(1) if total else None
                    })

        df = pd.DataFrame(extracted_data)

        if output_excel:
            df.to_excel(output_excel, index=False)

        print("Data extraction complete.")
        return df

    except Exception as e:
        print(f"An error occurred: {e}")
        return pd.DataFrame()

# Example usage
if __name__ == "__main__":
    pdf_path = "invoice.pdf"  # Replace with the path to your PDF
    output_excel = "invoice_data.xlsx"  # Replace with the desired output file path

    extracted_df = extract_invoice_data(pdf_path, output_excel)
    print(extracted_df)
```
